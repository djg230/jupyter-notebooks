```python
import random
import string


KEY_SPACE = string.ascii_uppercase


def validate_key(key):
    assert sorted(list(key.upper())) == list(KEY_SPACE)


def gen_key():
    k = list(KEY_SPACE)
    random.shuffle(k)
    return ''.join(k)


def encrypt(key, message):
    validate_key(key)

    cipher = ''
    for c in message:
        if c.upper() in key:
            i = KEY_SPACE.find(c.upper())
            x = key[i]
            cipher += x.upper() if c.isupper() else x.lower()
        else:
            cipher += c
    return cipher


def decrypt(key, cipher):
    validate_key(key)

    message = ''
    for c in cipher:
        if c.upper() in key:
            i = key.find(c.upper())
            x = KEY_SPACE[i]
            message += x.upper() if c.isupper() else x.lower()
        else:
            message += c
    return message


def main():
    # k = 'WZFEPYXDNCQRVUMTHBKOSILAGJ'
    #a    'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    k = gen_key()
    with open("y.txt", 'r') as f:
      pt = f.read().strip();
    ct = encrypt(k, pt)
    
    print (k)
    print (ct)


if __name__ == '__main__':
    main()
```
